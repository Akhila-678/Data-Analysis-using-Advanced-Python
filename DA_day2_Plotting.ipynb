Problem 2. Plot the scores of two students in 5 different subjects.
Subjects as x-axis and Marks as y-axis

```python
stu1=[56,78,92,95,78]
stu2=[77,89,100,73,45]
sub=["Maths","Science","English","Telugu","Social"]

plt.plot(sub,stu1,label="Student 1",color="pink")
plt.plot(sub,stu2,label="Student 2")
plt.legend()
```

```python
#sub-plot
stu1=[56,78,92,95,78]
stu2=[77,89,100,73,45]
sub=["Maths","Science","English","Telugu","Social"]

plt.subplot(2,1,1)
plt.plot(sub,stu1,label="Student 1",color="pink")
plt.legend()

plt.subplot(2,1,2)
plt.plot(sub,stu2,label="Student 2")
plt.legend()
```

Problem 3. Create Subplots(bar) showing the profits of two companies.
Given : Revenue of Company A and B in Lakhs
2019.   230   200
2020.   560   160
2020.   780   270
2020.   127   127
2020.   128   400



```python
#bar graph
com1 = np.array([230,560,780,127,128])
com2 = np.array([200,160,270,127,400])
print("Revenues of Company A",com1)
print("Revenues of Company B",com2)
years = ["19-20","20-21","21-22","22-23"]

c=np.diff(com1)
d=np.diff(com2)

print("Profits of A",c)
print("Profits of B",d)

plt.subplot(1,2,1)
plt.bar(years,c,label="Company1",color="k")
plt.legend(loc="best")

plt.subplot(1,2,2)
plt.bar(years,d,label="Company2",color="g")
plt.legend(loc="best")
```
